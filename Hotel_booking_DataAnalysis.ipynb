# Data Analysis and Visualization

```python
#finding the percentange of the cancelations
cancel_perc = data['is_canceled'].value_counts(normalize = True)
cancel_perc
```

0 represents not canceled and 1 represents cancel. we can see that the cancellation rate is half of the reservations

```python
plt.figure(figsize=(5,4))
plt.title("Reservation Rate count")
plt.bar(['Not Canceled', 'Canceled'], data['is_canceled'].value_counts(), edgecolor='k' , width= 0.7)
plt.show
```

Cancelled rate is almost half of the reservation which not so good for the company

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x= 'hotel' , hue = 'is_canceled' , data = data , palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
#ax1.legend(bbox_to_anchor(1,1))
plt.title("Reservation status in different hotels", size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
```

```python
resort_hotel = data[data['hotel'] =='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = data[data['hotel'] =='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

From this we can se that the cancellation rate is more in the city hotels. so we need to give more focus on the city hotels to reduce the cencellation rates

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

checking if the prices where the reason for the cancelation

```python
plt.figure(figsize = (20,8))
plt.title("Average daily rate of resort and city hotel")
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City hotel')
plt.legend()
plt.show()
```

```python
data['month'] = data['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1= sns.countplot(x='month' , hue ='is_canceled', data = data, palette = 'bright')
legend_labels,_ = ax1. get_legend_handles_labels()
#ax1.legend(bbox_to_anchor(1,1))
plt.title("Reservation status per month", size = 20)
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

january month has highest cancelation and august has lowest whereas August has highest reservation and Jan has lowest\

```python
plt.figure(figsize = (15,8))
plt.title("ADR per month", fontsize = 30)
sns.barplot('month','adr',data=data[data['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
canceled_data =data[data['is_canceled']==1]
top_10_country = canceled_data['country'].value_counts()[:10] #slicing for top 10 records
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with cancellations')
plt.pie(top_10_country, autopct ='%.2f' , labels = top_10_country.index)
plt.show()
```

Portugal countries has more rate of cancellation so company so include more facitly in the portugal countries hotels
